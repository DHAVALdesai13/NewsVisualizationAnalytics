```python
import spacy
from spacy import displacy
from collections import Counter
import en_core_web_sm
from pprint import pprint
import csv
import pandas as pd
import time
import pprint
import geonamescache
import json

nlp = en_core_web_sm.load()
```

```python
csv.field_size_limit(100000000)
csvpath = "all_sampled2.csv"
num_articles = 1e10

news = []

nb_per_source = 5000
news_sources = {'Reuters': nb_per_source, 'The New York Times': nb_per_source, 'People': nb_per_source, 'CNBC': nb_per_source, 'Vice': nb_per_source, 'The Hill': nb_per_source, 'CNN': nb_per_source, 'Mashable': nb_per_source}

with open(csvpath) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    t0 = time.time()
    nb_added_articles = 0
    for i, row in enumerate(csv_reader):
        
        if ((i % 100) == 0):
            t1 = time.time() - t0
            print("Time to parse {} articles: {}".format(i, t1))
            
        if (nb_added_articles == 8 * nb_per_source):
            print("Parsing completed.")
            break
        # if i%1000==0:
        #   print(i)
        if i>=num_articles:
            break
        article = ",".join(row[7:-3])
        date = row[2].strip().split(' ')[0].replace('-', '')[1:7]+"01"
        source = row[-1].split("'")[1]
        link = row[-3]
        
        news_source_dict = {}
        news_source_dict['date'] = date
        news_source_dict['news_source'] = source
        news_source_dict['url'] = link
        news_source_dict['article'] = article
        
        if (source not in news_sources):
            continue
        else:
            if (news_sources[source] <= 0):
                continue
            news_sources[source] = news_sources[source] - 1
            nb_added_articles += 1
        
        doc = nlp(article)
        
        locations = []
        for X in doc.ents:
            if (X.label_ == 'GPE'):
                locations.append(X.text)
                
        news_source_dict['locations'] = locations
        
        news.append(news_source_dict)
        
        if ((nb_added_articles % 100) == 0):
            print("{} articles were added.".format(nb_added_articles))
        
```

```python
#crossreferencing with the GeoNames databse
gc = geonamescache.GeonamesCache()

continents = gc.get_continents()
us_states = gc.get_us_states()
countries = gc.get_countries()
cities = gc.get_cities()

def gen_dict_extract(var, key):
    if isinstance(var, dict):
        for k, v in var.items():
            if k == key:
                yield v
            if isinstance(v, (dict, list)):
                yield from gen_dict_extract(v, key)
    elif isinstance(var, list):
        for d in var:
            yield from gen_dict_extract(d, key)
            
continents = [*gen_dict_extract(continents, 'name')]
us_states = [*gen_dict_extract(us_states, 'name')]
countries = [*gen_dict_extract(countries, 'name')]
cities = [*gen_dict_extract(cities, 'name')]
```

```python
#continuation of the crossreferencing, identifying cities, states, countries and continents

continents_l = []
us_states_l = []
countries_l = []
cities_l = []

for counter, new in enumerate(news):
    
    for location in new['locations']:
        if location in countries:
            if (location not in ['Georgia']):
                countries_l.append(location)
        if location in continents:
            continents_l.append(location)
        if location in us_states:
            us_states_l.append(location)
        if location in cities:
            if (location not in ['Florida','Virginia']):
                cities_l.append(location)
            
    #Removing duplicates
    continents_l = list(set(continents_l))
    us_states_l = list(set(us_states_l))
    countries_l = list(set(countries_l))
    cities_l = list(set(cities_l))
            
    news[counter]['continents'] = continents_l.copy()
    news[counter]['us_states'] = us_states_l.copy()
    news[counter]['countries'] = countries_l.copy()
    news[counter]['cities'] = cities_l.copy()
    
    continents_l.clear()
    us_states_l.clear()
    countries_l.clear()
    cities_l.clear()
    
```

```python
#aggregating data for every news source
blank_dictionary = {'countries': {}, 'cities': {}, 'continents': {}, 'us_states': {}}

aggregated_data = {'Reuters': {'countries': {}, 'cities': {}, 'continents': {}, 'us_states': {}}, 'The New York Times': {'countries': {}, 'cities': {}, 'continents': {}, 'us_states': {}}, 'People': {'countries': {}, 'cities': {}, 'continents': {}, 'us_states': {}}, 'CNBC': {'countries': {}, 'cities': {}, 'continents': {}, 'us_states': {}}, 'Vice': {'countries': {}, 'cities': {}, 'continents': {}, 'us_states': {}}, 'The Hill': {'countries': {}, 'cities': {}, 'continents': {}, 'us_states': {}}, 'CNN': {'countries': {}, 'cities': {}, 'continents': {}, 'us_states': {}}, 'Mashable': {'countries': {}, 'cities': {}, 'continents': {}, 'us_states': {}}}

for counter, new in enumerate(news):
    for country in new['countries']:
        if country in aggregated_data[new['news_source']]['countries']:
            aggregated_data[new['news_source']]['countries'][country] += 1
        else:
            aggregated_data[new['news_source']]['countries'][country] = 1
            
    for continent in new['continents']:
        if continent in aggregated_data[new['news_source']]['continents']:
            aggregated_data[new['news_source']]['continents'][continent] += 1
        else:
            aggregated_data[new['news_source']]['continents'][continent] = 1
            
    for us_state in new['us_states']:
        if us_state in aggregated_data[new['news_source']]['us_states']:
            aggregated_data[new['news_source']]['us_states'][us_state] += 1
        else:
            aggregated_data[new['news_source']]['us_states'][us_state] = 1
            
    for city in new['cities']:
        if city in aggregated_data[new['news_source']]['cities']:
            aggregated_data[new['news_source']]['cities'][city] += 1
        else:
            aggregated_data[new['news_source']]['cities'][city] = 1
```

```python
pprint.pprint(aggregated_data)
```

```python
aggregated_data_freq = aggregated_data.copy()
for news_source in aggregated_data:
    s = sum(aggregated_data[news_source]['cities'].values())
    for city in aggregated_data[news_source]['cities']:
        aggregated_data_freq[news_source]['cities'][city] = aggregated_data[news_source]['cities'][city] / s
        
    s = sum(aggregated_data[news_source]['continents'].values())
    for continent in aggregated_data[news_source]['continents']:
        aggregated_data_freq[news_source]['continents'][continent] = aggregated_data[news_source]['continents'][continent] / s
        
    s = sum(aggregated_data[news_source]['us_states'].values())
    for us_state in aggregated_data[news_source]['us_states']:
        aggregated_data_freq[news_source]['us_states'][us_state] = aggregated_data[news_source]['us_states'][us_state] / s
        
    s = sum(aggregated_data[news_source]['countries'].values())
    for country in aggregated_data[news_source]['countries']:
        aggregated_data_freq[news_source]['countries'][country] = aggregated_data[news_source]['countries'][country] / s
    
    
```

```python
nb_top_countries = 5
nb_top_continents = 1
nb_top_us_states = 5
nb_top_cities = 5

aggregated_data_freq_top = aggregated_data_freq.copy()
for news_source in aggregated_data_freq:
    s_cities= sorted(aggregated_data_freq[news_source]['cities'], key = aggregated_data_freq[news_source]['cities'].get, reverse = True)[:nb_top_cities]
    aggregated_data_freq_top[news_source]['top_cities'] = dict()
    for key in s_cities:
        aggregated_data_freq_top[news_source]['top_cities'][key] = aggregated_data_freq[news_source]['cities'][key]
        
    s_countries= sorted(aggregated_data_freq[news_source]['countries'], key = aggregated_data_freq[news_source]['countries'].get, reverse = True)[:nb_top_countries]
    aggregated_data_freq_top[news_source]['top_countries'] = dict()
    for key in s_countries:
        aggregated_data_freq_top[news_source]['top_countries'][key] = aggregated_data_freq[news_source]['countries'][key]
        
    s_continents= sorted(aggregated_data_freq[news_source]['continents'], key = aggregated_data_freq[news_source]['continents'].get, reverse = True)[:nb_top_continents]
    aggregated_data_freq_top[news_source]['top_continents'] = dict()
    for key in s_continents:
        aggregated_data_freq_top[news_source]['top_continents'][key] = aggregated_data_freq[news_source]['continents'][key]
        
    s_us_states= sorted(aggregated_data_freq[news_source]['us_states'], key = aggregated_data_freq[news_source]['us_states'].get, reverse = True)[:nb_top_us_states]
    aggregated_data_freq_top[news_source]['top_us_states'] = dict()
    for key in s_us_states:
        aggregated_data_freq_top[news_source]['top_us_states'][key] = aggregated_data_freq[news_source]['us_states'][key]
```

```python
top = {'Reuters': {'top_countries': {}, 'top_cities': {}, 'top_continents': {}, 'top_us_states': {}}, 'The New York Times': {'top_countries': {}, 'top_cities': {}, 'top_continents': {}, 'top_us_states': {}}, 'People': {'top_countries': {}, 'top_cities': {}, 'top_continents': {}, 'top_us_states': {}}, 'CNBC': {'top_countries': {}, 'top_cities': {}, 'top_continents': {}, 'top_us_states': {}}, 'Vice': {'top_countries': {}, 'top_cities': {}, 'top_continents': {}, 'top_us_states': {}}, 'The Hill': {'top_countries': {}, 'top_cities': {}, 'top_continents': {}, 'top_us_states': {}}, 'CNN': {'top_countries': {}, 'top_cities': {}, 'top_continents': {}, 'top_us_states': {}}, 'Mashable': {'top_countries': {}, 'top_cities': {}, 'top_continents': {}, 'top_us_states': {}}}

for news_source in aggregated_data_freq_top:
    for top_w in top[news_source]:
        top[news_source][top_w] = aggregated_data_freq_top[news_source][top_w].copy()
```
